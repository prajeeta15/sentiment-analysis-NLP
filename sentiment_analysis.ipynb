```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import requests
from bs4 import BeautifulSoup
import re
```

```python
tokenizer = AutoTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
model = AutoModelForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
```

```python
r = requests.get('https://www.yelp.com/biz/the-perch-chicago?osq=Restaurants') #request to the webpage to let us access the data on the webpage and gives you a response code
soup = BeautifulSoup(r.text, 'html.parser') #passed the response code from the above code and scrapes the webpage data to look for specifics
regex = re.compile('.*comment.*') #regex looks for anything starthing with comment in each class
results = soup.find_all('p',{'class':regex}) #this saves all the comments in class found by soup and p is for the paragraph tag
reviews = [result.text for result in results]
#scraping
```

```python
r.text #this is the response code that requests return and shows every text that is visible on the webpage and then this is passed through beautisul soup to be scraped and used
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.array(reviews), columns=['review'])
```

```python
df.head()
```

```python
df.tail()
```

```python
df['review'].iloc[0]
```

```python
def sentiment_score(review):
  tokens = tokenizer.encode(review,return_tensors='pt')
  result =  model(tokens)
  return int(torch.argmax(result.logits))+1
```

```python
sentiment_score(df['review'].iloc[1])
```

```python
#for mass review processing we use lambda
df['sentiment']=df['review'].apply(lambda x:sentiment_score(x[:600]))
```
